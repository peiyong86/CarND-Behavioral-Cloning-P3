```python
import os, csv
import numpy as np
```

```python
datapath = './alldata2/hardneg6/'
```

```python
imagelist = os.listdir(os.path.join(datapath, 'IMG'))
```

```python
labelfile = os.path.join(datapath, 'driving_log.csv')
angle = 1.5
```

```python
with open(labelfile, 'w') as f:
    writer = csv.writer(f)
    for imname in imagelist:
        line = [imname, 'x', 'x', str(angle + np.random.rand()/1.0)]
        writer.writerow(line)
```
